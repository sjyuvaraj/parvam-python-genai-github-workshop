```python
# Import pandas
# Used to work with DataFrames and CSV files
import pandas as pd
```

```python
# DataFrame() - Create DataFrame from dictionary
data = {"Name": ["A", "B", "C", "D"], "Age": [20, 30, 40, 50]}
df = pd.DataFrame(data)
df
```

```python
# DataFrame() - Create DataFrame from list of dictionaries
df = pd.DataFrame([
    {"Name": "Akshay", "Age": 24},
    {"Name": "Ajay", "Age": 23},
    {"Name": "Jay", "Age": 35},
])
df
```

```python
# read_csv() - Load data from CSV file
df = pd.read_csv("data.csv")
df.head()
```

```python
# info() - Display dataset structure and types
df.info()
```

```python
# head() - Show top 2 rows
df.head(2)
```

```python
# tail() - Show bottom 2 rows
df.tail(2)
```

```python
# Column Selection - Access specific column
df = pd.read_csv("weather.csv")
df["city"]
```

```python
# loc - Select row by label
df.loc[2]
```

```python
# iloc - Select rows by index
# Same as that of slicing, start is inclusive, end is exclusive
df.iloc[0:3]
```

```python
# Conditional Filtering - Filter rows based on condition
df[df["temp"] > 35.5]
```

```python
# Multiple Conditions - Apply AND condition
df[(df["humidity"] > 40) & (df["humidity"] < 50)]
```

```python
# Update Column - Modify column values
df["humidity"] = df["humidity"] + 1
df
```

```python
# isnull() - Check missing values
df = pd.read_csv("weather.csv")
df.isnull()
```

```python
# fillna() - Replace missing values
df.fillna(0, inplace=True)
df
```

```python
# describe() - Statistical summary of data
df.describe()
```

```python
# mean() - Calculate average of column
df["humidity"].mean()
```

```python
# sum() - Calculate total of column
df["humidity"].sum()
```

```python
df.sort_values("temp")
```

```python
# sort_values() - Sort data by column
df.sort_values("city")
```

```python
# groupby() - Group data and aggregate
data = {
    "Dept": ["Dev", "Dev", "QA", "Dev", "QA"],
    "Salary": [15000, 18000, 13000, 20000, 16000]
}
df = pd.DataFrame(data)
df.groupby("Dept").sum()
```

```python
# unique() - Get unique values
df["Dept"].unique()
```

```python
# value_counts() - Count occurrences
df["Dept"].value_counts()
```

```python
# to_csv() - Export DataFrame to CSV
df.to_csv("dept.csv", index=False)
```
