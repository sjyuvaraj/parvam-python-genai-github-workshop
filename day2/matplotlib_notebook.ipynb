```python
# Import matplotlib
# Used for plotting graphs and charts
import matplotlib.pyplot as plt
```

```python
# Basic Plot (Line Plot) - Draw a simple line graph
x = [1, 2, 3, 4]
y = [10, 15, 25, 20]
plt.plot(x, y)
plt.show()
```

```python
# Titles & Labels - Add title and axis labels
plt.plot(x, y)
plt.title("Sales Trend")
plt.xlabel("Days")
plt.ylabel("Sales")
plt.show()
```

```python
# Multiple Lines - Plot more than one line
y2 = [5, 15, 10, 22]
plt.plot(x, y)
plt.plot(x, y2)
plt.show()
```

```python
# Scatter Plot - Show data points as dots
plt.scatter(x, y)
plt.scatter(x, y2)
plt.show()
```

```python
# Bar Chart - Compare categories
categories = ["A", "B", "C"]
values = [10, 25, 18]
plt.title("Bar Chart for Categories & Values")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.bar(categories, values)
plt.show()
```

```python
# Histogram - Show frequency distribution
data = [10, 20, 20, 30, 30, 40, 40, 40]
plt.hist(data)
plt.show()
```

```python
# Pie Chart - Show percentage distribution
labels = ["A", "B", "C", "D", "E"]
sizes = [10, 50, 15, 10, 15]
plt.pie(sizes, labels=labels)
plt.show()
```

```python
# Grid - Add grid lines
plt.plot(x, y)
plt.grid()
plt.show()
```

```python
# Legend - Show labels for multiple lines
plt.plot(x, y, label="Sales")
plt.plot(x, y2, label="Profit")
plt.legend()
plt.show()
```

```python
# Figure Size - Adjust plot size
plt.figure(figsize=(10, 5))
plt.plot(x, y)
plt.show()
```

```python
# Subplots - Multiple graphs in one figure
plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.subplot(1, 2, 2)
plt.plot(x, y2)
plt.show()
```

```python
# Save Plot - Save chart as image
plt.plot(x, y)
plt.savefig("line-chart.png")
```

```python
# Line Style & Markers - Customize line appearance
plt.plot(x, y, linestyle="--", marker="o")
plt.show()
```

```python
# Box Plot - Show data distribution
data = [10, 20, 30, 40, 100]
plt.boxplot(data)
plt.show()
```
